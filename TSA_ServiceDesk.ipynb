```python
from statsmodels.tsa.stattools import adfuller
result=adfuller(y_train)
print(result)


# in ADF - Null hypothesis the series is not stationary , Alternate hypothesis is stationary

print(f'test stats: {result[0]}')
print(f'p-values:{result[1]}')
print(f'critical values:{result[4]}')

if result[1]>0.05:
  print('series is not stationary')
else:
  print('series is stationary')
```

```python
import itertools
p=d=q=range(0,3)

pdq=list(itertools.product(p,d,q))
seasonal_pdq=[(x[0],x[1],x[2],3)for x in list(itertools.product(p,d,q))]


print('examples of parameter combinations of Seasonal ARIMA')
print('SARIMAX:{}* {}'.format(pdq[1],seasonal_pdq[1]))
print('SARIMAX:{}* {}'.format(pdq[1],seasonal_pdq[2]))
print('SARIMAX:{}* {}'.format(pdq[2],seasonal_pdq[3]))
print('SARIMAX:{}* {}'.format(pdq[2],seasonal_pdq[4]))
print('SARIMAX:{}* {}'.format(pdq[3],seasonal_pdq[5]))



```

```python
metric_aic_dict=dict()

for pm in pdq:
    for pm_seasonal in seasonal_pdq:
        try:
            model=sm.tsa.statespace.SARIMAX(y_train,
                                           order=pm,
                                           seasonal_order=pm_seasonal,
                                           enforce_stationarity=False,
                                           enforce_invertibility=False)
            
            model_aic=model.fit()
            print('ARIMA{}*{}3 - AIC:{}'.format(pm,pm_seasonal,model_aic.aic))
            metric_aic_dict.update({(pm,pm_seasonal):model_aic.aic})
        except:
             continue
              
```

```python
{k:v for k,v in sorted(metric_aic_dict.items(),key=lambda x:x[1])}
```

```python
import statsmodels.api as sm
```

```python
#fitting the model:

model=sm.tsa.statespace.SARIMAX(y_train,
                               order=(0,0,0),
                               seasonal_order=(2,2,1,3),
                               enforce_stationarity=False,
                               enforce_invertibility=False)

model_aic=model.fit()
print(model_aic.summary().tables[1])
```

```python
import numpy as np
```

```python
forecast=model_aic.get_prediction(start=pd.to_datetime('2022-02-01'),dynamic=False)
predictions=forecast.predicted_mean

actual=y_test['2022-03-01':]

from sklearn.metrics import accuracy_score,mean_squared_error

rmse=np.sqrt((predictions-actual)**2).mean()

print('rmse value is {}'.format(round(rmse,2)))
```

```python
forecast=model_aic.get_forecast(steps=14)

predictions=forecast.predicted_mean
confidence_interval=forecast.conf_int()

fig=y.plot(label='actual',figsize=(14,7))
fig.set_xlabel('Month')
fig.set_ylabel('Orders')
fig.fill_between(confidence_interval.index,
                confidence_interval.iloc[:,0],
                confidence_interval.iloc[:,1],color='k',alpha=.2)


#prediction plot

predictions.plot(ax=fig,label='Predictions',alpha=.7,figsize=(14,7))

plt.legend()
plt.show()
```
